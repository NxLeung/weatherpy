```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as ow
from urllib.error import HTTPError

# Import API key
from api_keys import weatherKey

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
settings = {"units": "metric", "appid": weatherKey}
weather_data = []

for city in cities:
    batchcounter = 1
    try:
        response_data = ow.get_current(city, **settings)  
        if response_data['cod'] != 200:
            print (f"There is an error, going to sleep for a minute")
            time.sleep(1.01)
            batchcounter += 1
        else:    
            print(f"Processing Record {cities.index(city)+1} of Set {batchcounter} | {city}")
            weather_data.append(response_data)
        
    except HTTPError:
        print (f"City not found. Skipping...")

# weather_data
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
summary = ["name","clouds.all","sys.country","dt","main.humidity", "coord.lat", "coord.lon","main.temp_max" ,"wind.speed"]

data = [response(*summary) for response in weather_data]

weather_data = pd.DataFrame(data)
weather_data.columns = ['City','Cloudiness','Country','Date','Humidity', 'Lat','Lng','Max Temp', 'Wind Speed']
weather_data.to_csv("output/weather_data.csv", index=False)
weather_data.head()



```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
lat_temp = weather_data[['City','Lat', 'Max Temp']].set_index('City')
ax = lat_temp.plot(x='Lat', y='Max Temp',kind='Scatter', c='steelblue', marker="o", alpha=.75, edgecolors='black', linewidth=1, grid=True)
plt.title(f"City Latitude vs Max Temperature (06/05/2019)")
ax.set(xlabel='Latitude', ylabel='Max Temperature (C)')
plt.savefig("Output/Lat_Temp.png", bbox_inches="tight")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
lat_hum = weather_data[['City','Lat', 'Humidity']].set_index('City')
ax = lat_hum.plot(x='Lat', y='Humidity',kind='Scatter', c='steelblue', marker="o", alpha=.75, edgecolors='black', linewidth=1, grid=True)
plt.title(f"City Latitude vs Humidity (06/05/2019)")
ax.set(xlabel='Latitude', ylabel='Humidity (%)')
plt.savefig("Output/Lat_Hum.png", bbox_inches="tight")
plt.show()
```

```python
lat_cloud = weather_data[['City','Lat', 'Cloudiness']].set_index('City')
ax = lat_cloud.plot(x='Lat', y='Cloudiness',kind='Scatter', c='steelblue', marker="o", alpha=.75, edgecolors='black', linewidth=1, grid=True)
plt.title(f"City Latitude vs Cloudiness (06/05/2019)")
ax.set(xlabel='Latitude', ylabel='Cloudiness (%)')
plt.savefig("Output/Lat_Cloud.png", bbox_inches="tight")
plt.show()
```

```python
lat_wind = weather_data[['City','Lat', 'Wind Speed']].set_index('City')
ax = lat_wind.plot(x='Lat', y='Wind Speed',kind='Scatter', c='steelblue', marker="o", alpha=.75, edgecolors='black', linewidth=1, grid=True)
plt.title(f"City Latitude vs Wind Speed (06/05/2019)")
ax.set(xlabel='Latitude', ylabel='Wind Speed (mph)')
plt.savefig("Output/Lat_Wind.png", bbox_inches="tight")
plt.show()
```
